# Data Pre-processing for Multi-Label text classification

```python
# Importing required libraries
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv("data/data.csv")
df.head()
```

```python
# separate explanatory and dependent variables
X = df.iloc[:,1]
y = df.iloc[:,2:]
```

```python
# split for cross-validation (train-60%, validation 20% and test 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=123)

X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=123)
```

```python
# Check if the class proportions are maintained in the splits
display(y.apply(lambda x:  x.value_counts()/x.value_counts().sum()))
display(y_train.apply(lambda x:  x.value_counts()/x.value_counts().sum()))
display(y_test.apply(lambda x:  x.value_counts()/x.value_counts().sum()))
display(y_val.apply(lambda x:  x.value_counts()/x.value_counts().sum()))
```
